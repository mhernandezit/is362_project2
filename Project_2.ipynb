# First question - what is the average price of a private room in Manhattan?
Let's dig into the pandas dataframe to find apartments in Manhattan, then, Private rooms, and finally run an average of the price

```python
ma_data = airbnb_data[airbnb_data['neighbourhood_group'].str.contains("Manhattan")]
ma_pr_data = ma_data[ma_data['room_type'].str.contains("Private room")]
ma_pr_data
```

```python
ma_pr_data["price"].mean()
```

For my final dataset, I chose to examine the SAT scores of all New York High Schools in 2010.

```python
sat_data = pd.read_csv('2010_SAT__College_Board__School_Level_Results.csv')
sat_data
```

# What are the top 5 high schools in NYC ranked by SAT scores

```python
sat_data.sort_values('Critical Reading Mean', ascending = False).head(5)
```

As we can see from above, there are a number of schools that submitted non-numeric data.  Let's turn those 's' strings to NaN nulls, so we can properly run a sort to get our top scores.

```python
good_data = sat_data.replace('s', np.nan)
good_data
```

Once more, let's do the sort and we'll have our top scores.

```python
good_data.sort_values('Critical Reading Mean', ascending = False).head(5)
```

# These were the top 5 schools sorted by their SAT critical reading mean scores.
